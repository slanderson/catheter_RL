```python
fig, ((ax11, ax12, ax13), (ax21, ax22, ax23)) = plt.subplots(2, 3, figsize=(6, 4), sharex=False, sharey=False)
axes = ((ax11, ax12, ax13), (ax21, ax22, ax23))
abc.plot_shades(ax11, include_labels = False)
ukf.plot_shades(ax12, color = purple, include_labels = False)
mbc.plot_shades(ax13, color = grey, include_labels = False)

abc_r.plot_shades(ax21, include_labels = False)
ukf_r.plot_shades(ax22, color = purple, include_labels = False)
mbc_r.plot_shades(ax23, color = grey, include_labels = False)

axes_ticks = [-15, -10, -5, 0, 5, 10, 15]
axes_ticks_names = ['', '-10','', '0','', '10', '']

axes_ticks = [-15, 0, 15]
axes_ticks_names = ['-15','0','15']

plt.setp(axes, 
         xticks = axes_ticks, 
         xticklabels= axes_ticks_names,
         yticks= axes_ticks,
         yticklabels=axes_ticks_names)
# label = axes.yaxis.get_major_ticks()[2].label
# label.set_fontsize(10)

sns.despine()
for ax in (ax11, ax12, ax13, ax21, ax22, ax23):
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xlim((-20, 22))
    ax.set_ylim((-19, 19))
for ax in (ax11, ax12, ax13):
    ax.spines['bottom'].set_visible(False)
    ax.xaxis.set_ticks([])
for ax in (ax12, ax13, ax22, ax23):
    ax.spines['left'].set_visible(False)
    ax.yaxis.set_ticks([])


    
left  = 0.2  # the left side of the subplots of the figure
right = 0.9    # the right side of the subplots of the figure
bottom = 0.2   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.0   # the amount of width reserved for blank space between subplots
hspace = -.15   # the amount of height reserved for white space between subplots

plt.subplots_adjust(left, bottom, right, top, wspace, hspace)
# plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=-0.0, hspace=-0.15)

ax22.set_xlabel(r'\textbf{Y (mm)}', fontsize = 16)

plt.sca(ax11)
rotation=90.
x = -43
y = -12
plt.text(x,
         y,
         r'\textbf{Z (mm)}',
         rotation=90.,
         fontsize = 16)

# fig.legend((),(),bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
legend_height = 1.3
blue_patch = mpatches.Patch(color=blue, alpha = 0.1)
blue_line,  = plt.plot([100, 100], [100, 100], color = blue)
legend_abc = plt.legend(handles=[(blue_patch, blue_line)], 
                        labels=[r'$\alpha \beta \gamma$'],
                        loc = 'upper left', 
                        bbox_to_anchor = [0,legend_height])
ax11.add_artist(legend_abc)

purple_patch = mpatches.Patch(color=purple, alpha = 0.1)
purple_line,  = plt.plot([100, 100], [100, 100], color = purple)
legend_ukf = plt.legend(handles=[(purple_patch, purple_line)], 
                        labels=['UKF'],
                        loc = 'upper left', 
                        bbox_to_anchor = [1.0,legend_height])
ax11.add_artist(legend_ukf)

grey_patch = mpatches.Patch(color=grey, alpha = 0.1)
grey_line,  = plt.plot([100, 100], [100, 100], color = grey)
legend_mbc = plt.legend(handles=[(grey_patch, grey_line)], 
                        labels=['Model'], 
                        loc = 'upper left', 
                        bbox_to_anchor = [2.0,legend_height])
ax11.add_artist(legend_mbc)

# fig.savefig('test.png')
# fig.savefig('test.pdf')

# fig.tight_layout()
# fig.legend(loc = 'upper left', bbox_to_anchor = [0.2,0.45])
```
